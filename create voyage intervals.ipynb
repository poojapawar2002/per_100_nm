```python
import pandas as pd
import os

# Set the folder path where your files are
folder_path = "../Data/inputfiles/"

# List all CSV files
csv_files = [f for f in os.listdir(folder_path) if f.startswith('autolog_input_with_dist')]
print("csv files ", csv_files )
df_list = []

for file in csv_files:
    file_path = os.path.join(folder_path, file)
    
    # Read CSV and drop fully empty rows
    df = pd.read_csv(file_path, skip_blank_lines=True)
    df = df.dropna(how='all')  # ðŸ’¡ Remove rows where all values are NaN

    df_list.append(df)

# Concatenate all dataframes
combined_df = pd.concat(df_list, ignore_index=True)

# Print shape to verify
print(f"Combined DataFrame shape: {combined_df.shape}")
combined_df.rename(columns = {"EndDateUTC_x": "EndDateUTC"}, inplace=True)
combined_df.drop(columns = ["EndDateUTC_y"], inplace=True)    
# Save to a new file
combined_df.to_csv("../Data/combined_input_files_7000series_with_dist.csv", index=False)
```

```python
import pandas as pd
df = voyage

# Sort by IMO and timestamp
df = df.sort_values(['IMO', 'StartDateUTC']).reset_index(drop=True)

# Initialize the dictionary
voyage_dict = {}

# Group by vessel
for imo, group in df.groupby('IMO'):
    group = group.sort_values('StartDateUTC').reset_index(drop=True)
    voyages = []
    voyage_start = False
    voyage_startdate = None
    voyage_enddate = None

    for _, row in group.iterrows():
        log = row['LogCategory']
        start_time = row['StartDateUTC']
        end_time = row["LogDateUTC"]

        if log == 'BOSP':
            voyage_start = True
        elif log == 'Sea Noon':
            if (voyage_startdate is None) and (voyage_start):
                voyage_startdate = start_time
            voyage_enddate = end_time
        elif log == 'EOSP' and voyage_start:
            if voyage_startdate is not None: 
                voyages.append((voyage_startdate, end_time))
            else:
                voyages.append((start_time, end_time))
            voyage_startdate = None 
            voyage_start = False
        elif log == "ARRIVAL" and voyage_start:
            if voyage_startdate is not None:
                voyages.append((voyage_startdate, voyage_enddate))
            voyage_startdate = None 
            voyage_start = False

    voyage_dict[imo] = voyages


for imo, vlist in voyage_dict.items():
    print(f"IMO: {imo}")
    for start, end in vlist:
        print(f"  Voyage: {start} â†’ {end}")
```

```python
import pickle

with open('voyage_dict.pkl', 'wb') as f:
    pickle.dump(voyage_dict, f)
```
